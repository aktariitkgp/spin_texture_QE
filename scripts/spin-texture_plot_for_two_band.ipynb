```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load your Excel file
file_path = 'kx_ky_sx_sy_sz_440_441.xlsx'
df = pd.read_excel(file_path)

# Extract the data for both bands (440 and 441)
# Assuming the columns follow a consistent structure: kx, ky, sx, sy, sz
kx_440 = df['kx_nbnd_440']
ky_440 = df['ky_nbnd_440']
sx_440 = df['sx_nbnd_440']
sy_440 = df['sy_nbnd_440']
sz_440 = df['sz_nbnd_440']

kx_441 = df['kx_nbnd_441']
ky_441 = df['ky_nbnd_441']
sx_441 = df['sx_nbnd_441']
sy_441 = df['sy_nbnd_441']
sz_441 = df['sz_nbnd_441']

# Plot for Band 440
plt.figure(figsize=(10, 8))
plt.quiver(kx_440, ky_440, sx_440, sy_440, sz_440, cmap='coolwarm', scale=10)
plt.colorbar(label='$s_z$ (Colorbar)')
plt.title('Band 440: Vector Field in $k_x, k_y$ Plane')
plt.xlabel('$k_x$')
plt.ylabel('$k_y$')
plt.grid(True)

# Show the plot for Band 440
plt.show()

# Plot for Band 441
plt.figure(figsize=(10, 8))
plt.quiver(kx_441, ky_441, sx_441, sy_441, sz_441, cmap='coolwarm', scale=10)
plt.colorbar(label='$s_z$ (Colorbar)')
plt.title('Band 441: Vector Field in $k_x, k_y$ Plane')
plt.xlabel('$k_x$')
plt.ylabel('$k_y$')
plt.grid(True)

# Show the plot for Band 441
plt.show()
```
