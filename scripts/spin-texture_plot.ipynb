```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load your Excel file
file_path = 'kx_ky_sx_sy_sz_440_441.xlsx'
df = pd.read_excel(file_path)

# Extract the data for both bands (440 and 441)
kx_440 = df['kx_nbnd_440']
ky_440 = df['ky_nbnd_440']
sx_440 = df['sx_nbnd_440']
sy_440 = df['sy_nbnd_440']
sz_440 = df['sz_nbnd_440']

kx_441 = df['kx_nbnd_441']
ky_441 = df['ky_nbnd_441']
sx_441 = df['sx_nbnd_441']
sy_441 = df['sy_nbnd_441']
sz_441 = df['sz_nbnd_441']

# Combine the data for both bands
kx_combined = np.concatenate((kx_440, kx_441))
ky_combined = np.concatenate((ky_440, ky_441))
sx_combined = np.concatenate((sx_440, sx_441))
sy_combined = np.concatenate((sy_440, sy_441))
sz_combined = np.concatenate((sz_440, sz_441))

# Create a DataFrame for the combined data
combined_data = pd.DataFrame({
    'kx_combined': kx_combined,
    'ky_combined': ky_combined,
    'sx_combined': sx_combined,
    'sy_combined': sy_combined,
    'sz_combined': sz_combined
})

# Save the combined data to a CSV file
combined_data.to_csv('combined_kx_ky_sx_sy_sz.csv', index=False)

# Plot combined data for both bands
plt.figure(figsize=(10, 8))
quiver = plt.quiver(kx_combined, ky_combined, sx_combined, sy_combined, sz_combined, 
                    cmap='coolwarm', scale=10)
plt.colorbar(quiver, label='$s_z$ (Colorbar)')
plt.title('Bands 440 & 441: Vector Field in $k_x, k_y$ Plane')
plt.xlabel('$k_x$')
plt.ylabel('$k_y$')
plt.grid(True)

# Show the plot
plt.show()
```
